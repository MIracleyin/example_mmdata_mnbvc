```python
def calculate_sha1(url):

    # # 创建SHA-1哈希对象
    # sha1 = hashlib.sha1()
    # # 更新哈希对象，需要将字符串编码为字节
    # sha1.update(text.encode())
    # # 获取16进制格式的哈希值
    # sha1_digest = sha1.hexdigest()
    hash = SHA1.new(url.encode("utf-8")).hexdigest()
    return hash
len(calculate_sha1('test'))
```

```python
from typing import Dict

class mmblock:
    def __init__(self, **kwargs) -> None:
        self.entity_id = kwargs.get("entity_id") 
        self.block_id = kwargs.get("block_id")
        self.timestamp = kwargs.get("timestamp")
        self.metadata = kwargs.get("metadata")
        self.text = kwargs.get("text")
        self.image = kwargs.get("image")
        self.ocr_text = kwargs.get("ocr_text")
        self.audio = kwargs.get("audio")
        self.stt_text = kwargs.get("stt_text")
        # self.other_block = kwargs.get("other_block") # ]
        self.block_type = kwargs.get("block_type")
        self.file_md5 = kwargs.get("file_md5")
        self.page_id = kwargs.get("page_id")
    
    def to_pydict(self) -> Dict:
        return {
            "实体ID": self.entity_id,
            "块ID": self.block_id,
            "时间": self.timestamp,
            "扩展字段": self.metadata,
            "文本": self.text,
            "图片": self.image,
            "OCR文本": self.ocr_text,
            "音频": self.audio,
            "STT文本": self.stt_text,
            "块类型": self.block_type,
            "文件md5": self.file_md5,
            "页ID": self.page_id,
        }
```

```python
# 文本信息处理
text = "这是一条 demo 数据"
text_meta_dict = {"文本长度": len(text)}
```

```python
# 图片信息处理
import io
import numpy as np
from PIL import Image

def random_gen_image(image_path):
    image = np.random.rand(200, 200, 3)
    image = (image * 255).astype(np.uint8)
    im = Image.fromarray(image)
    im.save(image_path)

def read_image(image_path):
    # 打开图像文件
    with open(image_path, 'rb') as file:
        image = Image.open(file)
        # 将图像转换为二进制格式
        img_byte_arr = io.BytesIO()
        image.save(img_byte_arr, format=image.format)
        img_byte_arr = img_byte_arr.getvalue()
    return img_byte_arr
random_gen_image("blocks/demo.png")
image = read_image("blocks/demo.png")
image_meta_dict = {"宽度": 200, "高度": 200}
image
```

```python
# 音频信息处理
from pydub import AudioSegment
from scipy.io.wavfile import write
def random_gen_audio(audio_path):
    sample_rate = 44100
    duration = 5.0
    # 生成随机音频
    samples = np.random.uniform(low=-1, high=1, size=int(sample_rate * duration))
    samples = np.int16(samples * 32767)
    # 保存为WAV文件
    write(audio_path, sample_rate, samples)
def read_audio(audio_path):
    # 使用PyDub读取音频文件
    audio = AudioSegment.from_wav(audio_path,)
    # 创建一个BytesIO对象
    byte_arr = io.BytesIO()
    # 将音频保存为二进制格式
    audio.export(byte_arr, format="wav")
    # 获取二进制数据
    audio_bytes = byte_arr.getvalue()

    return audio_bytes

random_gen_audio("blocks/demo.wav")
audio = read_audio("blocks/demo.wav")
audio_meta_dict = {"长度": 5, "采样率": 44100}
audio
```

```python
timestamp = datetime.now()
formatted_time = timestamp.strftime("%Y%m%d")

entity_id = "mnbvc_mm_example"
timestamp = formatted_time

block_type = "文本" # 和当前块保存信息的列名一致
file_md5 = "542adfbadc34e2a1e975a2c8df6058ca"
page_id = None
demo_blocks = [
    mmblock(
        entity_id=entity_id,
        block_id=0,
        timestamp=formatted_time,
        metadata=json.dumps(text_meta_dict, ensure_ascii=False),
        text=text,
        file_md5=file_md5 # todo
    ),
    mmblock(
        entity_id=entity_id,
        block_id=1,
        timestamp=formatted_time,
        metadata=json.dumps(image_meta_dict, ensure_ascii=False),
        image=image,
        file_md5=file_md5, # todo
    ),
    mmblock(
        entity_id=entity_id,
        block_id=2,
        timestamp=formatted_time,
        metadata=json.dumps(audio_meta_dict, ensure_ascii=False),
        audio=audio,
        file_md5=file_md5, # todo
    )
]
[block.to_pydict() for block in demo_blocks]
```
